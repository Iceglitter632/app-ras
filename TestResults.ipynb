```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('results/1st_draft/train_loss.txt') as f:
    train_loss = [float(line[:-2]) for line in f]

with open('results/1st_draft/test_loss.txt') as f:
    test_loss = [float(line[:-2]) for line in f]
```

```python
train_loss = [i if i < 1 else np.mean(train_loss) for i in train_loss]
```

```python
plt.plot(train_loss, label = "train")
plt.plot(test_loss, label = "test")
plt.xlabel("steps")
plt.ylabel("loss")
plt.legend()
plt.show()
```
